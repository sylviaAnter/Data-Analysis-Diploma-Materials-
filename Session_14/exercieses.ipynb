# 10- you have list of numbers create 2 list one for +ve , -ve

#11- you have a dic of names and grades of student create 2 list one for names , grades

````python
dic={
    'names':['aya','noha','mai','nesriin'],
    'grades':[100,90,80,70]
}
names=[]
grades=[]
for i in dic:
  if i == 'names':
    names.append(dic[i])
  else:
    grades.append(dic[i])


print(names)
print(grades)

````
