#Make Login System

```python
def login(user_name,password):
  if user_name=="Ola" and password=="123":
    return True
  else:
    return False
```

```python
while True:
  user_name=input("Enter user name: ")
  password=input("Enter password: ")
  Login_condition=login(user_name,password)
  if Login_condition==True:
    print("Welcome",user_name)
    print("Login Done !")
    break
  else:
    continue
```
