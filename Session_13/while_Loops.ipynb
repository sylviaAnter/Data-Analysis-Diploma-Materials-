# while Loops



```python
x = 0
while x <= 10:
    print('x is currently: ',x)
    print(' x is still less than 10, adding 1 to x')
    x+=1  #x=x+1   #x++
```

#Keeps asking the user to enter numbers
#Stops when the user enters 0  
#Prints the sum of all entered numbers

```python
num=int(input("pls enter the number (if zero App Stop)"))
total=0

while num !=0:
    total+=num
    num=int(input("pls enter the number (if zero App Stop)"))


print("the sum of all entered numbers ",total)
```

#Write a program that:Takes a number from the user
#Prints a countdown from that number to 0

```python
num=int(input("plz enter the number"))
while num >=0:
  print(num)
  num-=1
```
