### Multiple Branches



```python
place = 'Bank'

if place == 'Auto Shop':
    print('Welcome to the Auto Shop!')
elif place == 'Bank':
    print('Welcome to the bank!')
else:
    print('Where are you?')
```

```python
person = 'Samy'

if person == 'SaMy':
    print('Welcome Sammy!')
else:
    print("Welcome, what's your name?")

```

###x > 85 (A+)
###x (75-85) -- (B+)
###x (65-75) -- C+
###x (50 - 65) (D+)
###x < 50  (F)

```python

x=int(input("Enter Ur Grades "))

if x>=85 and x<=100:
  print("A+")
elif x>=75 and x <85:
  print("B+")
elif x>=65 and x<75:
  print("C+")
elif x>=50 and x <65:
  print("D+")
elif x<50:
  print("F")
else:
  print("Enter valid number")
```
